<p>Les vraquiers sont des navires qui permettent le transport de marchandises solides en vrac : blé, orge, soja,  charbon, sable … Cette matière première est stockée dans des cales qui sont remplies et vidées souvent par des grues portuaires ou des tapis roulant. Ces cales sont fermées par des portes appelées écoutilles. </p>

<p>Il existe différents types de navires vraquiers. La coque se défini en général à partir de ce modèle.</p>



<p>il existe différents moyens de charger le vraquier</p>



<p>Structure d’un vraquier : 
Le ratio longueur/largeur est compris entre 	5 et 7, pour une moyenne de 6,2<br>
Le ratio Longueur/creux est compris entre 11 et 12<br>
Les écoutilles font entre 45% et 60% de la largeur du pont et entre 57% et 67% de la longueur des cales
</p>
